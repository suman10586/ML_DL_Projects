```python
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
data = pd.read_csv('lsd-math-score-data.csv')
```

```python
data.describe()
```

```python
data.isnull().any()
```

```python
type(data)
```

```python
LSD = pd.DataFrame(data , columns = ['LSD_ppm'])
score = pd.DataFrame(data , columns = ['Avg_Math_Test_Score'])
time =  pd.DataFrame(data , columns = ['Time_Delay_in_Minutes'])
```

```python
regr = LinearRegression()
regr.fit(LSD,score)
```

```python
regr.intercept_
```

```python
regr.coef_
```

```python
predicted_score = regr.predict(LSD)
```

```python
plt.title('Concentration of LSD ppm with Time', fontsize=17)
plt.xlabel('Time Delay in Minutes ')
plt.ylabel('LSD Concentration in PPM')
plt.text(x = -0.2, y = -0.2, s='Wagner et al. (1986)')
plt.plot(time,LSD, color='orange', lw=3)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.style.use('classic')
```

```python
plt.title('Airthematic VS LSD_25', fontsize=17)
plt.ylabel('Average Maths Scores ')
plt.xlabel('LSD Concentration in PPM')
plt.scatter(LSD,score,color ='navy',s = 100)
plt.style.use('seaborn')
plt.plot(LSD,predicted_score, color='red',lw=2)
plt.xlim(0,8)
plt.ylim(0,90)
plt.show()
```
